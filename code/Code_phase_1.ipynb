# Imports and function definitions

```python
%matplotlib inline 
import os
import sys
import dicom
import matplotlib.pyplot as plt
from dicom.errors import InvalidDicomError

import numpy as np
import pandas as pd
from PIL import Image, ImageDraw

import logging
root = logging.getLogger()
root.addHandler(logging.StreamHandler())
for handler in root.handlers[:]:
    root.removeHandler(handler)
logging.basicConfig(format = u'[%(asctime)s]  %(message)s', level = logging.INFO)


def parse_contour_file(filename):
    """Parse the given contour filename

    :param filename: filepath to the contourfile to parse
    :return: list of tuples holding x, y coordinates of the contour
    """

    coords_lst = []

    with open(filename, 'r') as infile:
        for line in infile:
            coords = line.strip().split()

            x_coord = float(coords[0])
            y_coord = float(coords[1])
            coords_lst.append((x_coord, y_coord))

    return coords_lst


def parse_dicom_file(filename):
    """Parse the given DICOM filename

    :param filename: filepath to the DICOM file to parse
    :return: dictionary with DICOM image data
    """

    try:
        dcm = dicom.read_file(filename)
        dcm_image = dcm.pixel_array

        try:
            intercept = dcm.RescaleIntercept
        except AttributeError:
            intercept = 0.0
        try:
            slope = dcm.RescaleSlope
        except AttributeError:
            slope = 0.0

        if intercept != 0.0 and slope != 0.0:
            dcm_image = dcm_image*slope + intercept
        dcm_dict = {'pixel_data' : dcm_image}
        return dcm_dict
    except InvalidDicomError:
        return None


def poly_to_mask(polygon, width, height):
    """Convert polygon to mask

    :param polygon: list of pairs of x, y coords [(x1, y1), (x2, y2), ...]
     in units of pixels
    :param width: scalar image width
    :param height: scalar image height
    :return: Boolean mask of shape (height, width)
    """

    # http://stackoverflow.com/a/3732128/1410871
    img = Image.new(mode='L', size=(width, height), color=0)
    ImageDraw.Draw(img).polygon(xy=polygon, outline=0, fill=1)
    mask = np.array(img).astype(bool)
    return mask
```

# Visualization

As not all DICOM images have contours (the number of contours is less than number of DICOM images) we need to figure out how to match contour with right .dcm file. Let's take a look on several i-contour file names:

*IM-0001-0048-icontour-manual.txt*

*IM-0001-0059-icontour-manual.txt*

*IM-0001-0068-icontour-manual.txt*

So each time the second number is changing and if we take the file with max number, it's always less than the max number for DICOM files for specific patient

That's why we will use code __int(f.split("-")[2])__ to get the right number for .dcm file. Let's take a look on some examples of reading DICOM images for specific patient and masking the i-contour (red region). 

```python
contour_path = '../input/contourfiles/SC-HF-I-1/i-contours/'
dicom_path = '../input/dicoms/SCD0000101/'
fig = plt.figure(figsize=(20,70))
for num, f in enumerate(sorted(os.listdir(contour_path))):
    cont = parse_contour_file(contour_path + f)
    #Example of contour name = IM-0001-0048-icontour-manual.txt
    img = parse_dicom_file(dicom_path + str(int(f.split("-")[2])) + '.dcm')
    mask = poly_to_mask(cont, img['pixel_data'].shape[0], img['pixel_data'].shape[1])
    y = fig.add_subplot(9,2,num+1)
    y.imshow(img['pixel_data'], cmap=plt.cm.gray)
    y.imshow(mask, 'jet', alpha=0.3)
plt.show()
```

Nice! That works as we thought so we can continue and do the same for each patient:

```python
dicom_path = '../input/dicoms/{}/'
contour_path = '../input/contourfiles/{}/i-contours/'

def create_train_dataset(dicom_path, contour_path, patient_to_contour):
    train_images = []
    Y_train = []
    for patient, contour in patient_to_contour:   
        d_path = dicom_path.format(patient)
        c_path = contour_path.format(contour)
        logging.info("Patient: {}, Contours: {}, Contours_num: {}".format(patient, contour, len(os.listdir(c_path))))
        for num, f in enumerate(sorted(os.listdir(c_path))):
            cont = parse_contour_file(c_path + f)
            img = parse_dicom_file(d_path + str(int(f.split("-")[2])) + '.dcm')
            mask = poly_to_mask(cont, img['pixel_data'].shape[0], img['pixel_data'].shape[1])
            train_images.append(img['pixel_data'])
            Y_train.append(mask.astype(int))
        logging.info("Current len train: {}, current len ytrain: {}".format(len(train_images), len(Y_train)))
    logging.info("Finished creation of training dataset")
    return np.array(train_images), np.array(Y_train)

train_images, Y_train = create_train_dataset(dicom_path, contour_path, patient_to_contour)
```

Now we have 2 variables - __train_images__ and __Y_train__ with train data and targets. Let's check once again and plot 4 random images with their target masks:

```python
fig = plt.figure(figsize=(20,20))
for num, f in enumerate([13, 77, 21, 35]):
    img = train_images[f]
    mask = Y_train[f]
    y = fig.add_subplot(2,2,num+1)
    y.imshow(img, cmap=plt.cm.gray)
    y.imshow(mask, 'jet', alpha=0.3)
plt.show()
```

Ok, now we are 100% sure that everything works fine. __If we want to use this code for production, the only thing that we need to change is adding new data to the input folder in the same manner as we have it now and update link.csv file - everything else will be done for us by create_training_dataset function__

Ok, to do that first of all we need to make a permutation of the dataset to fulfill the third item of this list. After that it will be possible to take 0 to 7 image (8 observations), then 8 to 15 (again 8 observations), etc.

```python
np.random.seed(13)
perm = np.random.permutation(len(Y_train))
# Apply to both labels and images
train_images = train_images[perm]
Y_train = Y_train[perm]
```

Checking once again that everything is fine with matching between labels and images:

As we can see, now we have really different images from those we get in part 1, but the red region is still on the place it should be. Now we can continue with batch generating:

```python
BATCH_SIZE = 8
for it, i in enumerate(range(0, len(Y_train), BATCH_SIZE)):
    logging.info("Preparing batch {}".format(it))
    X_batch = train_images[i:i + BATCH_SIZE]
    Y_batch = Y_train[i:i + BATCH_SIZE]
    logging.info("Batch {} is ready".format(it))
```

If we want to do several epochs on that dataset, the code should look like this:

```python
np.random.seed(13)
NUM_EPOCH = 2
for epoch in range(NUM_EPOCH):
    perm = np.random.permutation(len(Y_train))
    logging.info("Permutation first 8: {}".format(perm[0:8]))
    train_images = train_images[perm]
    Y_train = Y_train[perm]
    for it, i in enumerate(range(0, len(Y_train), BATCH_SIZE)):
        logging.info("Preparing batch {} for epoch {}".format(it, epoch))
        X_batch = train_images[i:i + BATCH_SIZE]
        Y_batch = Y_train[i:i + BATCH_SIZE]
        logging.info("Batch {} is ready".format(it))
```
